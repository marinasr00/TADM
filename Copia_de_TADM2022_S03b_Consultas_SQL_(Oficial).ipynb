## Group By

1. ¿Cuántos préstamos hay por cada estado y cual es la media de la cantidad de dinero prestado?

*Nota*:

* where: se filtran los registros antes de agruparlos
* having: se filtran los grupos tras obtenerlos

La media: avg()

```python
%%sql
SELECT status, count (*) as total, avg (amount)
FROM LOAN
WHERE duration >= 24
GROUP BY status
ORDER BY total DESC
```

De los grupos quiero aquellos que del total únicamente salgan 25 o al menos 25 préstamos:


```python
%%sql
SELECT status, count (*) as total, avg (amount)
FROM LOAN
WHERE duration >= 24
GROUP BY status
HAVING total > 25
ORDER BY total DESC
```

Se busca la tabla CLIENT, porque es donde existe esta variable.

2º Posteriormente, cuando sabemos como se llama, la meto en la tabla de abajo, en este caso, sería birth_number.

```python
%%sql
SELECT loan_id, A4, A9, A10, A11, c.birth_number, l.date loan_date, amount, duration, payments, frequency, a.date acc_date, status
FROM LOAN l, ACCOUNT a, DISTRICT d, DISPOSITION dp, CLIENT c
WHERE l.account_id = a.account_id
AND a.district_id = d.A1
AND dp.account_id = a.account_id 
AND dp.client_id = c.client_id
AND status in ("A", "B")
ORDER BY amount DESC, a.date DESC
LIMIT 10
```

Despues, se hará un sumatorio de las variables:

```python
%%sql
SELECT count (*) loan_id, A4, A9, A10, A11, c.birth_number, l.date loan_date, amount, duration, payments, frequency, a.date acc_date, status
FROM LOAN l, ACCOUNT a, DISTRICT d, DISPOSITION dp, CLIENT c
WHERE l.account_id = a.account_id
AND a.district_id = d.A1
AND dp.account_id = a.account_id -- join disposition y account
AND dp.client_id = c.client_id -- join disposition y client
AND status in ("A", "B")
ORDER BY amount DESC, a.date DESC
LIMIT 10
```

14. Prestamos cuya cantidad es superior a la media de los clientes de Praga

- Cantidad de préstamos existentes en cuentas de Praga 

```python

%%sql
SELECT count (*)
FROM LOAN L, ACCOUNT a, DISTRICT D
WHERE l.account_id = a.account_id
AND a.district_id = d.A1 
AND d.A3 = "Prague"
```

Tengo 84 préstamos de personas que viven en Praga.

* Cantidad media prestada a los cuentas de Praga

```python

%%sql
SELECT avg(l.amount)
FROM LOAN L, ACCOUNT a, DISTRICT D
WHERE l.account_id = a.account_id
AND a.district_id = d.A1 
AND d.A3 = "Prague"
```

A continuación, se consultarán el total de los préstamos que tengan una cantidad superior a la media de los clientes:

```python
%%sql
SELECT count(*)
FROM LOAN
WHERE amount > (SELECT avg(l.amount)
        FROM LOAN L, ACCOUNT a, DISTRICT D
        WHERE l.account_id = a.account_id
        AND a.district_id = d.A1 
        AND d.A3 = "Prague")
```

En este caso, 277 préstamos superan la media.

Ahora, se van a consultar los préstamos que tengan una cantidad superior a la media de los clientes:

Extra: se limita a 10.

```python
%%sql
SELECT * FROM LOAN 
WHERE amount > (SELECT avg(l.amount)
        FROM LOAN l, ACCOUNT a, DISTRICT d
        WHERE l.account_id = a.account_id and a.district_id = d.A1  
        and d.A3 = "Prague")
ORDER BY amount DESC
LIMIT 10
```

15 . Clientes de Praga con préstamos según su estado

- ¿cuántas de esas cuentas tienen asociados un préstamo?
- Necesito saber cuántos hay por cada estado (*status*)


```python
%%sql 
SELECT status, count(*)
FROM CLIENT c, DISTRICT d, DISPOSITION dp, LOAN l 
WHERE c.district_id = d.A1 and c.client_id = dp.client_id
and l.account_id = dp.account_id and d.A3 = "Prague"
GROUP BY status
```

## Ejercicio

> Cantidad de dinero medio en transacciones de clientes propietarios de cuentas en Kolin (owned accounts)

```python
%%sql
SELECT c.client_id, avg (amount) as av
from client c, district d, disposition ds, account a, trans t
WHERE c.district_id = d.A1 and c.client_id = ds.client_id
    and ds.account_id = a.account_id and a.account_id = t.account_id
and d.A2 = "Kolin" and ds.type = "OWNER" 
GROUP BY c.client_id
ORDER BY av DESC
LIMIT 10

```
